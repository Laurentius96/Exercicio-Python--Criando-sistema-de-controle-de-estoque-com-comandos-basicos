## 1. Criando um mini sistema de controle de estoque

- Crie um sistema para ser usado pelo time de controle de estoque de um centro de distribuição.
- Imagine que ao fim de todo dia, o time conta quantas unidades de produto existem no estoque. Se tivermos um estoque abaixo do estoque permitido para aquela categoria do produto, o time deve ser avisado (print) para fazer um novo pedido daquele produto.
- Cada categoria de produto tem um estoque mínimo diferente, segundo a regra abaixo:

- alimentos -> Estoque mínimo: 50
- bebidas -> Estoque mínimo: 75
- limpeza -> Estoque mínimo: 30

Para isso vamos criar um programa que pede 3 inputs do usuário: nome do produto, categoria e quantidade atual em estoque.

Se o produto tiver abaixo do estoque mínimo da categoria dele, o programa deve printar a mensagem "Solicitar {produto} à equipe de compras, temos apenas {unidades} em estoque"

Exemplo: Se o usuário preenche os inputs com: bebidas, dolly, 90, o programa não deve exibir nenhuma mensagem.<br>
Agora, se o usuário preenche os inputs com: bebidas, guaraná, 60, o programa deve exibir a mensagem "Solicitar guaraná à equipe de compras, temos apenas 60 unidades em estoque.

Obs: lembre de usar o int() para transformar o número inserido pelo usuário no input de string para int.<br>
Obs2: Caso o usuário não preencha alguma das 3 informações, o programa deve exibir uma mensagem para avisá-lo de preencher corretamente.

```python
# Variáveis:
    # nome_p ----> Nome do Produdo
    # categoria -> Categoria do Produto
    # qtde ------> Quantidade do Produto em estoque
    # verif -----> verificação 

nome_p = input('Informe a marca do produto: ')

categoria = input('Informe a categoria do produto - Alimentos, Bebidas ou Limpeza: ')

qtde = input('Informe a quantidade em estoque do produto: ' )

verif = qtde.isdigit() # qtde.isdigit() retorna True se a string tem apenas números.

alimento = 50
           
bebidas = 75
           
limpeza = 30


# Verificar se deixou em branco...
if nome_p and categoria and qtde: 
    
    # If para Alimento...
    if 'limento' in categoria: # O 'limento' com in permite ao usuário escrever alimento ou Alimento e no plural ou singular.
        
        if verif == False : # Verificando se preencheu com uma letra a quantidade do produto.  
            
            print('Erro: Colocou letra ao informar a quantidade do produto!')   
    
        
        elif int(qtde) > alimento: # Verificando se o estoque atual é maior que seu estoque minimo.
           
            print('Atualmente, o estoque para mínimo para Alimentos é: {}' .format(alimento))
            print('O produto {} está acima do estoque mínimo!' .format(nome_p))
        
    
        elif int(qtde) == alimento: # Verificando se o estoque atual é igual ao seu estoque minimo.
            
            print('Atualmente, o estoque para mínimo para Alimentos é: {}' .format(alimento))
            print('O produto {} chegou ao estoque mínimo, preparar o setor de compras!' .format(nome_p))
            
        
        else: # Verificando se o estoque atual é igual ao seu estoque minimo.     
            
            print('Atualmente, o estoque para mínimo para Alimentos é: {}' .format(alimento))
            print('Solicitar {} à equipe de compras, temos apenas {} em estoque!' .format(nome_p,qtde))
            
      

 
  #-------------------------------           
    # If para Bebidas...        
    elif 'ebida' in categoria: # O 'ebida' com in permite ao usuário escrever bebidas ou Bebidas e no plural ou singular.
        
        if verif == False : # Verificando se preencheu com uma letra a quantidade do produto.  
            
            print('Erro: Colocou letra ao informar a quantidade do produto!') 

        
        elif int(qtde) > bebidas: # Verificando se o estoque atual é maior que seu estoque minimo.
           
            print('Atualmente, o estoque para mínimo para Bebidas é: {}' .format(alimento))
            print('O produto {} está acima do estoque mínimo!' .format(nome_p))
        
      
        elif int(qtde) == bebidas: # Verificando se o estoque atual é igual ao seu estoque minimo.
            
            print('Atualmente, o estoque mínimo para Bebidas é: {}' .format(alimento))
            print('O produto {} chegou ao estoque mínimo, preparar o setor de compras!' .format(nome_p))
               
      
        else: # Verificando se o estoque atual é igual ao seu estoque minimo.     
            
            print('Atualmente, o estoque para mínimo para Bebidas é: {}' .format(alimento))
            print('Solicitar {} à equipe de compras, temos apenas {} em estoque!' .format(nome_p,qtde))


   #-------------------------------           
    # If para Limpeza...        
    elif 'impeza' in categoria: # O 'impeza' com in permite ao usuário escrever limpeza ou Limpeza e no plural ou singular.
         
        if verif == False : # Verificando se preencheu com uma letra a quantidade do produto.  
            
            print('Erro: Colocou letra ao informar a quantidade do produto!') 

                
        elif int(qtde) > limpeza: # Verificando se o estoque atual é maior que seu estoque minimo.
           
            print('Atualmente, o estoque para mínimo para Limpeza é: {}' .format(alimento))
            print('O produto {} está acima do estoque mínimo!' .format(nome_p))
        
        
        elif int(qtde) == limpeza: # Verificando se o estoque atual é igual ao seu estoque minimo.
            
            print('Atualmente, o estoque mínimo para Limpeza é: {}' .format(alimento))
            print('O produto {} chegou ao estoque mínimo, preparar o setor de compras!' .format(nome_p))           
                   
        
        else: # Verificando se o estoque atual é igual ao seu estoque minimo.     
            
            print('Atualmente, o estoque para mínimo para Limpeza é: {}' .format(alimento))
            print('Solicitar {} à equipe de compras, temos apenas {} em estoque!' .format(nome_p,qtde))    
            
        
            

#Se algum dos inputs estiverem vazios então...
else:

    if nome_p == '': #Verifica se o imput vazio é o nome_produto

        print('Erro: deixou o nome do produto em branco!')

    if categoria == '': #Verifica se o imput vazio é o categoria

        print('Erro: deixou a categoria do produto em branco!')

    if qtde == '': #Verifica se o imput vazio é o qntd_produto

        print('Erro: deixou a quantidade do produto em branco!')
```
